```python
import pandas as pd
```

```python
SourceData = pd.read_feather(
    "/mnt/upramdya_data/MD/MultiMazeRecorder/Datasets/Skeleton_TNT/241218_FinalEventCutoffData_norm/contact_data/231115_TNT_Fine_1_Videos_Tracked_contact_data.feather"
)
```

```python
SourceData.head()
```

```python
# Take 3 random flies and get how many contacts they have

# Pick 3 values in the range of the number of flies
# Randomly pick 3 flies

Flies = SourceData["fly"].unique()

import random

random.seed(42)

RandomFlies = random.sample(list(Flies), 3)

RandomFlies
```

```python
# For each fly, get the number of contacts they have

for fly in RandomFlies:
    data = SourceData[SourceData["fly"] == fly]

    print(f"Fly {fly} has {data["contact_index"].max()} contacts")
```
