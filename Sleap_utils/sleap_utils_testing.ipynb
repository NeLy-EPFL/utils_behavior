```python
# Find the first frame in the fly tracks where one of the legs is at less than 13 pixels from the ball

FlyPos = TestTracks.dataset

FlyPos
```

```python
# Merge Ball dataset to FlyPos on the frame number
FlyPos = FlyPos.merge(balltracks.dataset, on="frame")

FlyPos
```

```python
# Compute the distance between the ball and the legs

import numpy as np

FlyPos["LFront_ball_dist"] = np.sqrt(
    (FlyPos["x_Lfront"] - FlyPos["x_centre"]) ** 2
    + (FlyPos["y_Lfront"] - FlyPos["y_centre"]) ** 2
)

FlyPos["LFront_ball_dist"]
```

```python
# Plot the distance between the ball (x y) and the fly Rfront and Lfront legs, colored by time
plt.plot(FlyPos["LFront_ball_dist"])
```

```python
import cv2
```

```python
# Find all consecutive frames where the distance is less than 13 pixels

Contacts = FlyPos[
    (FlyPos["LFront_ball_dist"] < 14)
    & (FlyPos["LFront_ball_dist"].shift(-1) < 14)
]

Contacts
```

```python
# Filter out the contacts that are too short
Contacts_Events = Contacts[Contacts["frame"].diff() > 29]

Contacts_Events
```

```python
# Plot the frames where the fly is close to the ball

for frame in Contacts_Events["frame"]:
    img = TestTracks.generate_annotated_frame(frame)
    img = cv2.circle(img, (int(balltracks.centre[frame][0]), int(balltracks.centre[frame][1])), 11, (0, 0, 255), 2)
    plt.imshow(img)
    plt.show()
```

```python
from sleap import Labels, Video
from sleap.nn.tracking import track_poses

def assign_tracks_to_slp(slp_file_path, output_file_path, tracker_type="flow"):
    # Load the existing SLP file
    labels = sleap.load_file(slp_file_path)
    
    # Get the video from the labels
    video = labels.videos[0]
    
    # Run the tracking process
    new_labels = track_poses(labels, tracker_type=tracker_type)
    
    # Save the updated SLP file with assigned tracks
    new_labels.save(output_file_path)
    print(f"Tracks assigned and saved to {output_file_path}")

# Example usage
slp_file_path = "/mnt/upramdya_data/MD/F1_Tracks/Videos/240924_F1_3mm_ends_Videos_Checked/arena2/Right/arena2_right_rotated_tracked_ball.slp"
output_file_path = "/mnt/upramdya_data/MD/F1_Tracks/Videos/240924_F1_3mm_ends_Videos_Checked/arena2/Right/arena2_right_rotated_tracked_ball_updated.slp"
assign_tracks_to_slp(slp_file_path, output_file_path)
```
