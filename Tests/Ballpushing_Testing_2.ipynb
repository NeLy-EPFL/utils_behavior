```python
# Plot the data

import seaborn as sns

plt.figure(figsize=(20, 10))

sns.scatterplot(
    x= "contact_index",
    y= "ball_displacement",
    data=Dataset.data,
)

plt.show()
```

```python
# DO the same but with a line and confidence interval

plt.figure(figsize=(20, 10))

sns.lineplot(
    x= "contact_index",
    y= "ball_displacement",
    data=Dataset.data,
    hue= "Nickname"
)

plt.show()
```

```python
# Find out where is the highest ball displacement per Nickname

# Get the index of the maximum ball displacement for each Nickname
max_displacement_idx = Dataset.data.groupby("Nickname")["ball_displacement"].idxmax()

# Use the index to get the corresponding contact_index
max_displacement_contact_indices = Dataset.data.loc[
    max_displacement_idx, ["Nickname", "contact_index", "ball_displacement"]
]

max_displacement_contact_indices
```

```python
# DO the same but look at only the 15 first contacts

plt.figure(figsize=(20, 10))

sns.lineplot(
    x= "contact_index",
    y= "ball_displacement",
    data=Dataset.data[Dataset.data["contact_index"] < 15],
)

plt.show()
```

```python
# Let's make one plot per fly

# Determine the global minimum and maximum values of ball_displacement
global_min = Dataset.data["ball_displacement"].min()
global_max = Dataset.data["ball_displacement"].max()

# Create one plot per fly with fixed scales for both axes
for fly in Dataset.data["fly"].unique():
    plt.figure(figsize=(20, 10))
    sns.lineplot(
        x="contact_index",
        y="ball_displacement",
        data=Dataset.data[
            (Dataset.data["fly"] == fly) & (Dataset.data["contact_index"] <= 25)
        ],
    )
    
    plt.ylim(global_min, global_max)  # Set fixed y-axis scale
    plt.xlim(0, 25)  # Set fixed x-axis scale
    plt.xticks(range(0, 26, 1))  # Set x-axis ticks with a step of 1
    plt.title(fly)
    plt.xlabel("Contact Index")
    plt.ylabel("Ball Displacement")
    plt.show()
```
